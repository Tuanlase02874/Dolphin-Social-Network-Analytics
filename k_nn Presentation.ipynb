# KNN Algorithm

```python
def predict_knn(G,S,k):
    size = S.shape[1]
    P = np.zeros((size,size))
    for i in range(size):
        for j in range(size):
            if i!= j and G.item(i,j) != 1.0:
                #find top k similar of d_j and compute predict 
                d_j =  S[:][j]
                index = d_j.argsort()[-k:][::-1]
                top_d_j = d_j[d_j.argsort()[-k:][::-1]]
                print top_d_j
                print index
                Connect = G[:][j].T
                print Connect
                P[i][j] = top_d_j*Connect[index]/sum(top_d_j)
            else:
                P[i][j] = -1
            
    return P

Predict = predict_knn(G,S,3)
print Predict
```
